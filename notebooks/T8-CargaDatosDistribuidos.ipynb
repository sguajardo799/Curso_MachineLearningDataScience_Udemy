## Unir datasets por filas
+ Axis 0 -> Horizontal o por filas
+ Axis 1 -> Vertical o por columnas

```python
# Añadimos nueva caracteristica
blanco = ["White" for _ in range(df_white.shape[0])]
df_white = df_white.assign(color=blanco)
df_white.head()
```

```python
rojo = ["Red" for _ in range(df_white.shape[0])]
df_red = df_white.assign(color=rojo)
df_red.head()
```

Ahora podemos combinar ambos dataset sin perder la informacion de que color de vino es el detallado

```python
df_wine = pd.concat([df_red, df_white], axis=0)
df_wine.shape
```

```python
# en el repo del curso ubicamos distributed-data
dd_path = "../resources/python-ml-course/datasets/distributed-data/"

# Revisamos como exploracion primer dato
data_demo = pd.read_csv(dd_path+"001.csv")
data_demo.head()
```

Comprobando su estructura obtenemos los nombres de los archivos en el fichero, validamos que sea un csv(con os.path.sliptext o pathlib), y almacenamos los nombres en una lista

```python
import os
from os.path import isfile, join, splitext
files_name = [f for f in os.listdir(dd_path) if isfile(join(dd_path,f)) and splitext(f)[1]==".csv"]
files_name
```

```python
len(files_name)
```

Ahora en un bucle recorremos el listado y vamos cargando y concatenando los datasets

```python
df_full = pd.DataFrame()
for f in files_name:
    df_temp = pd.read_csv(dd_path+f)
    df_full = pd.concat([df_full,df_temp], axis=0)

df_full.shape
```

Lo anterior lo podemos convertir en una funcion como:

```python
def df_from_directory(dir:str, type_file:str = "csv"):
    df_full = pd.DataFrame()
    files_name = [f for f in os.listdir(dir) if isfile(join(dir,f)) and splitext(f)[1]=="."+type_file]
    for f in files_name:
        df_temp = pd.read_csv(dir+f)
        df_full = pd.concat([df_full,df_temp], axis=0)

    return df_full

df_sensors = df_from_directory(dir=dd_path)
df_sensors.head()
```
