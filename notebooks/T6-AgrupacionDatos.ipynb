# Agrupacion de Datos
+ Normalmente para agrupar datos se emplean variables categoricas
+ Es util para hacer analisis categoricos

```python
import numpy as np
import pandas as pd
```

## Dummy Dataset
+ Datasets de pruebas generados a partir de distribuciones probabilisticas

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500 # numero de datos
gender_data = []
income_data = []

for i in range(n):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
def generate_income(income:list):
    income_d = []
    for i in range(len(income)):
        if income[i] == "Poor":
            income_d.append(4000 + 3500*np.random.randn())
        elif income[i] == "Middle Class":
            income_d.append(12000 + 3500*np.random.randn())
        elif income[i] == "Rich":
            income_d.append(18000 + 3500*np.random.randn())
        else:
            income_d.append(0)
    return income_d
```

```python
height = 160 + 30*np.random.randn(n)
weight = 65 + 25*np.random.randn(n)
age = 30 + 12*np.random.randn(n)
income = generate_income(income_data)
```

```python
data = pd.DataFrame({
    "Gender": gender_data,
    "Economic Status": income_data,
    "Height": height,
    "Weight": weight,
    "Age": age,
    "Income": income
})

data.head()
```

## Agrupacion de Datos - Group By

```python
grouped_gender = data.groupby("Gender")
grouped_gender.groups
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
grouped_gender.get_group("Female")
```

```python
double_group = data.groupby(["Gender", "Economic Status"])
for names, groups in double_group:
    print(names)
    print(groups)
```
